# Tensors

Dit voorbeeld gaat uit van onderstaand neural network. 2 inputs (x1 & x2). en 2 output neuronen (a1 & a2).

Input waardes:
x1 = 1

x2 = 2

Ook zetten we de gewichten vast op de volgende waardes:

w1 = 1; 
w2 = 2; 
w3 = 3; 
w4 = 4; 

b1 = 1;
b2 = 2;


afbeelding.png

```python
import torch
import torch.nn as nn

inputs = torch.tensor([[1.0,2.0]])

model = nn.Linear(2, 2)

model.weight.data[0][0] = 1 #w1
model.weight.data[0][1] = 2 #w2
model.weight.data[1][0] = 3 #w3
model.weight.data[1][1] = 4 #w4

model.bias.data[0] = 1
model.bias.data[1] = 2

model_params = list(model.parameters())

weights = model_params[0].data
bias = model_params[1].data

print("weights:")
print(weights)


print("\nweights transposed:")
print(weights.T)

print("\nbias:")
print(bias)
```

afbeelding.png

De gewichten worden door pytorch als eerste geïnitieerd in een (2,2) matrix. Echter wordt deze transposed om de forward pass te maken.

Stel de output laag waren niet 2 neuronen maar 3. Dan was dit een (2,3) matrix na de transpose:

afbeelding.png

# De forward pass

Uiteindelijk voeren we het model uit om de forward pass uit te voeren:

a1 = (1x1) + (2x2) + 1 = 6

a2 = (1x3) + (2x4) + 2 = 13

#We kunnen dit ook in numpy nabootsen:

```python
import numpy as np

input = np.array([1,2])
weights = np.array([[1,2],
                    [3,4]])
bias = np.array([1,2])
print("Weights transposed")
print(weights.T)

outputs = np.dot(input, weights.T) + bias
print("\nOutput")
print(outputs)
```
